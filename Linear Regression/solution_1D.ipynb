```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading data
X = []
Y = []
for line in open('data_1d.csv'):
    x, y = line.split(',')
    X.append(float(x))
    Y.append(float(y))
```

```python
#converting to numpy arrays
X = np.array(X)
Y = np.array(Y)
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
#apply equations to calculate a and b in 'ax+b'
denom = X.dot(X) - X.mean()* X.sum()
a = (Y.dot(X) - Y.mean() * X.sum())/denom
b = (Y.mean() * X.dot(X) - X.mean() * Y.dot(X))/denom
```

```python
yhat = a*X + b
plt.scatter(X,Y)
plt.plot(X, yhat, color='r')
plt.show()
```

```python
# finding r**2 error
d1 = Y - yhat
d2  = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
r2
```
