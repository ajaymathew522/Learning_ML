```python
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = []
Y = []
non_decimal = re.compile(r'[^\d+]')
```

```python
for line in open('moore.csv'):
    r = line.split('\t')
    x = int(non_decimal.sub('',r[2].split('[')[0]))
    y = int(non_decimal.sub('',r[1].split('[')[0]))
    X.append(x)
    Y.append(y)
```

```python
X = np.array(X)
Y = np.array(Y)
plt.scatter(X,Y)
plt.show()
```

```python
 Y = np.log(Y)
 plt.scatter(X,Y)
 plt.show()
```

```python
denom = X.dot(X) - X.mean()* X.sum()
a = (Y.dot(X) - Y.mean() * X.sum())/denom
b = (Y.mean() * X.dot(X) - X.mean() * Y.dot(X))/denom
```

```python
yhat = a*X + b
plt.scatter(X,Y)
plt.plot(X, yhat, color='r')
plt.show()
```

```python
d1 = Y - yhat
d2  = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
r2
```
